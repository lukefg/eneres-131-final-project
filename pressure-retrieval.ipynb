```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/hurdat2_cleaned.csv")
```

```python
df.rename(axis=1, mapper={'min_pressure_mb':'central_pressure_mb'}, inplace=True)
```

```python
gb = df.groupby('code').agg({'central_pressure_mb': 'min'})
```

```python
gb = gb.reset_index()
```

```python
gb.columns = ['code','min_pressure_mb']
```

```python
gb.to_csv('data/pressure.csv', index=False)
```
