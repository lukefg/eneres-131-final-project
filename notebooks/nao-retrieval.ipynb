```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
formation = pd.read_csv('../data/processed/predictors/formation.csv', parse_dates=['formation_datetime'])
nao = pd.read_csv('../data/raw/nao.tsv', sep='\t', index_col=0)
nao.head()
```

```python
def get_nao(dt):
	result = nao.loc[dt.year, str(dt.month)]
	return np.nan if result == -99.99 else result
```

```python
formation['nao'] = formation.apply(lambda row: get_nao(row.formation_datetime), axis=1)
```

```python
results = formation.drop(columns=['formation_datetime', 'formation_lat', 'formation_lon'])
results
```

```python
results.to_csv('../data/processed/predictors/nao.csv', index=False)
```
