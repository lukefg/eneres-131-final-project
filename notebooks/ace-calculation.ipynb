```python
import numpy as np
import pandas as pd
import netCDF4 as nc
from datetime import date, timedelta
import os
import xarray as xr
```

```python
df = pd.read_csv("../data/raw/hurdat2_cleaned.csv", parse_dates=['formation_datetime', 'observation_datetime'])
```

```python
def calc_summand(v):
	return v**2 if v >= 34 else 0
```

```python
df['vmax_squared_gt_32'] = df.apply(func = lambda row: calc_summand(row['max_sus_wind_kt']), axis=1)
df.head()
```

```python
gb = df.groupby('code').agg({'vmax_squared_gt_32': 'sum'})
```

```python
gb.vmax_squared_gt_32 = gb.vmax_squared_gt_32 / 10**4
gb
```

```python
gb.columns = ['ace']
gb
```

```python
gb.reset_index(inplace = True)
```

```python
gb.to_csv('data/ace.csv', index=False)
```
