```python
import numpy as np
import pandas as pd
import netCDF4 as nc
from datetime import date, timedelta
import os
import xarray as xr
```

```python
hurdat = pd.read_csv("data/hurdat2_cleaned.csv", parse_dates=['formation_datetime', 'observation_datetime'])
```

```python
hurdat_sorted = hurdat.sort_values(by=['code','observation_datetime'], ascending=[True, True]).drop_duplicates(keep='first', subset=['code'])
```

```python
hurdat_sorted.sort_values(by=['observation_datetime'], ascending=[True], inplace=True)
```

```python
columns_to_keep = ['code','name','formation_datetime','lat','lon']
df = hurdat_sorted[columns_to_keep].copy()
df.columns = ['code','name','formation_datetime','formation_lat','formation_lon']
df.head()
```

```python
df['year'] = df['formation_datetime'].dt.year
df['dayofyear'] = df['formation_datetime'].dt.dayofyear
```

```python
features = ['ace','rhum','sst','vws']
for feature in features:
	data = pd.read_csv(f"data/{feature}.csv")
	df = df.merge(data, on='code', how='left')
```
